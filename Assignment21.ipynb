Q1. What is a probability distribution, exactly? If the values are meant to be random, how can you predict them at all?

A probability distribution is a mathematical function or description that defines the likelihood of different outcomes or values in a random experiment or process. It provides a way to model and quantify uncertainty and randomness.

In probability theory, there are two main types of probability distributions:

1. **Discrete Probability Distribution**: This type of distribution is used when the random variable can only take on distinct, separate values. Examples include the Bernoulli distribution (for binary outcomes like coin flips) and the Poisson distribution (for counting events in a fixed interval).

2. **Continuous Probability Distribution**: Continuous distributions are used when the random variable can take on any value within a specified range. The normal distribution (bell curve) and the exponential distribution are examples of continuous probability distributions.


Q2. Is there a distinction between true random numbers and pseudo-random numbers, if there is one? Why are the latter considered “good enough”?

Yes, there is a distinction. True random numbers are generated from inherently unpredictable physical processes, while pseudo-random numbers are generated by deterministic algorithms. Pseudo-random numbers are considered "good enough" for many applications because they exhibit statistical randomness properties, making them suitable for simulations, cryptography, and most computational tasks. They are computationally efficient and reproducible, but not suitable for applications requiring absolute unpredictability.

Q3. What are the two main factors that influence the behaviour of a "normal" probability distribution?

The two main factors influencing the behavior of a "normal" (Gaussian) probability distribution are:

1. **Mean (μ)**: The mean represents the central value or average of the distribution. It determines the center of the bell-shaped curve.

2. **Standard Deviation (σ)**: The standard deviation measures the spread or dispersion of data points around the mean. A higher standard deviation results in a wider distribution.


Q4. Provide a real-life example of a normal distribution.

A real-life example of a normal distribution is the distribution of adult human heights in a population. In this scenario, heights tend to cluster around the mean height, forming a bell-shaped curve. The mean height represents the average, and heights deviating slightly above or below the mean are more common than extreme deviations, following the characteristics of a normal distribution.

Q5. In the short term, how can you expect a probability distribution to behave? What do you think will happen as the number of trials grows?

In the short term, the behavior of a probability distribution can be unpredictable, as individual trials may yield varied results. However, as the number of trials grows, the distribution tends to converge to its expected or theoretical shape. This is a fundamental principle of probability theory known as the Law of Large Numbers, which states that with more trials, the observed outcomes tend to match the theoretical probabilities more closely, reducing short-term variability.

Q6. What kind of object can be shuffled by using random.shuffle?

The `random.shuffle` function in Python can be used to shuffle objects that are sequences or collections, such as lists, tuples, and arrays. It rearranges the elements within the object in a random order, effectively changing their positions. However, it cannot be used with immutable sequences like strings or with non-sequence objects like integers or floats, as they lack the required sequence structure.

Q8. What is the relationship between exponentiation and logarithms?

Exponentiation and logarithms are inverse mathematical operations that are related in the following way:

1. **Exponentiation**: In exponentiation, you raise a base number to a certain power to get a result. For example, 2^3 = 8, where 2 is the base and 3 is the exponent.

2. **Logarithms**: Logarithms are the inverse of exponentiation. They tell you what power you need to raise a specific base to in order to get a certain result. In the example above, log base 2 of 8 is 3, which means 2^3 = 8.

Mathematically, this relationship is expressed as:

- Exponentiation: a^b = c
- Logarithms: log_a(c) = b

Q9. What are the three logarithmic functions that Python supports?

Python supports three main logarithmic functions:

1. **Natural Logarithm (ln)**: The `math.log()` function without a specified base calculates the natural logarithm (base e) of a number.

2. **Common Logarithm (log10)**: The `math.log10()` function calculates the base-10 logarithm of a number, also known as the common logarithm.

3. **Custom Base Logarithm**: You can calculate the logarithm with a custom base (e.g., base 2) using `math.log(x, base)`. For base 2, it's often used for binary-related calculations.
